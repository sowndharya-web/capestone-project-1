#KERALA BUSES ROUTES-1

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
wait = WebDriverWait(driver,30)

driver.get('https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile')
driver.maximize_window()

def kerala_route_links(xpath):
    links_kerala = []
    routes_kerala = []

    for i in range(1, 4):  
        elements = driver.find_elements(By.XPATH, xpath)
        for element in elements:
           
                href = element.get_attribute("href")
                links_kerala.append(href)
        for route in elements:
                routes_kerala.append(route.text)
          
        # Page navigation
        try:
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            print("Pagination container found.")

           
            next_button_xpath= f'//div[@class="DC_117_pageTabs " and text()={i+1}]'
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH,next_button_xpath)))
            print(f"Next button found for page {i+1}: {next_button_xpath}")
           
            # Scroll into view and click
            driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
            time.sleep(1)  # Allow scrolling to complete
            actions = ActionChains(driver)
            actions.move_to_element(next_button).click().perform()

            # Wait for the page to load
            time.sleep(5)
        except (NoSuchElementException, TimeoutException):
            print(f"No more pages to paginate after page {i}")
            break

    return links_kerala, routes_kerala

# Update the XPath to match the correct elements on the webpage
links_kerala, routes_kerala = kerala_route_links("//a[@class='route']")

# Print the results
print("Links:")
for link in links_kerala:
    print(link)

print("\nRoutes:")
for route in routes_kerala:
    print(route)

# Close the WebDriver
driver.quit()
```

```python
df_k=pd.DataFrame({"Route_name":routes_kerala,"Route_links":links_kerala})
df_k
```

```python
df_k.to_csv('kerala_routes.csv', index=False)
# Print a message to confirm saving
print("DataFrame saved to 'kerala_routes.csv'")
```
