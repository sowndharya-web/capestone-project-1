```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
import pandas as pd
from selenium.common.exceptions import NoSuchElementException, TimeoutException
```

KERALA BUS DETAILS-1

```python
#reding the csv file:
df=pd.read_csv("kerala_routes.csv")
df
```

```python
#retriving the bus details:
driver_k=webdriver.Chrome()
Bus_name_k=[]
Bus_type_k=[]
Start_time_k=[]
End_time_k=[]
Time_duration_k=[]
Rating_k=[]
Prices_k=[]
Routes_k=[]
Links_k=[]
Seat_Availability_k=[]

#iterating through each rows :
for i,r in df.iterrows():
    route=r["Route_name"]
    link= r["Route_links"]  
    #looping through each link:
    driver_k.get(link)
    time.sleep(2)
    #clicking on the element to retive the bus details:
    elements=driver_k.find_elements(By.XPATH,f"//a[contains(@href,'{link}')]")
    '''for element in elements:
        element.click()
        time.sleep(2)'''
    #click the element to view the bus details:
    for element in elements:
        try:  
            link_click = wait.until(EC.element_to_be_clickable((By.XPATH,f"//a[contains(@href,'{link}')]")))
            actions = ActionChains(driver_k)
            actions.move_to_element(link_click).click().perform()
          
        except:
            continue
    
    time.sleep(5)
    #scrolling to end of the page :
    scrolling=True
    while scrolling:
        old_page_source=driver_k.page_source
        ActionChains(driver_k).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(5)
        new_page_source=driver_k.page_source
        if new_page_source==old_page_source:
            scrolling=False

    #extracting thebus details:
    bus_name_elements=driver_k.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")
    bus_type_elements=driver_k.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_time_elements=driver_k.find_elements(By.XPATH,"//div[@class='dp-time f-19 d-color f-bold']")
    end_time_elements=driver_k.find_elements(By.XPATH,"//div[@class='bp-time f-19 d-color disp-Inline']")
    time_duration_elements=driver_k.find_elements(By.XPATH,"//div[@class='dur l-color lh-24']")
    rating_elements=driver_k.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    price_elements=driver_k.find_elements(By.XPATH,"//div[@class='fare d-block']")
 
    seat_availability_elements =driver_k.find_elements(By.XPATH,"//div[@class='seat-left m-top-30']")
   
# Ensure that the number of items in each list is the same
    max_len = max(len(bus_name_elements), len(bus_type_elements), len(start_time_elements),
                  len(end_time_elements), len(time_duration_elements), len(rating_elements),
                  len(price_elements), len(seat_availability_elements))
    
    for idx in range(max_len):
        Bus_name_k.append(bus_name_elements[idx].text if idx < len(bus_name_elements) else None)
        Bus_type_k.append(bus_type_elements[idx].text if idx < len(bus_type_elements) else None)
        Start_time_k.append(start_time_elements[idx].text if idx < len(start_time_elements) else None)
        End_time_k.append(end_time_elements[idx].text if idx < len(end_time_elements) else None)
        Time_duration_k.append(time_duration_elements[idx].text if idx < len(time_duration_elements) else None)
        Rating_k.append(rating_elements[idx].text if idx < len(rating_elements) else None)
        Prices_k.append(price_elements[idx].text if idx < len(price_elements) else None)
        Seat_Availability_k.append(seat_availability_elements[idx].text if idx < len(seat_availability_elements) else None)
        Routes_k.append(route)
        Links_k.append(link)

print("Completed")

# Close the WebDriver
driver_k.quit()
            
                

        
        
        
```
